```python
%%writefile program.l

%{
    int count=0;
%}

%%
[A-Za-z_][A-Za-z_]*   { count++; }
.|\n                  ;
%%

int yywrap(void) {
    return 1;
}

int main(int argc, char *argv[]) {
    if(argc!=2){
        printf("Usage:\n\t./a.out \n");
        exit(0);
    }
    yyin=fopen(argv[1],"r");

    yylex();
    printf("number of identifiers = %d\n", count);

    fclose(yyin);
    return 0;
}
```

```python
%%writefile program.txt
123
num
num2
2num
123xyz
_123
_num
num_123
123_num
```

```python
%%shell

lex -l program.l
gcc lex.yy.c
./a.out program.txt
```
